```python
%load_ext autoreload
%autoreload 2
    
import typing
from typing import List, Iterable
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from utils import *

import os, glob
```

```python
data_ls = []
for path in glob.glob("dataset/*/rf_results.pkl"):
    with open(path,"rb") as file:
        data = pickle.load(file)
        data_ls.append(data)
```

```python
# hit recovery rate / percentage screened
for d in data_ls:
    plt.plot(d['percentage_iter'], d['recovery_iter'], c=(0.8,0.8,0.8))
plt.plot(np.mean([d['percentage_iter'] for d in data_ls], axis=0), np.mean([d['recovery_iter'] for d in data_ls], axis=0), c=(0,0,0), label='mean')
plt.plot(np.mean([d['percentage_iter'] for d in data_ls], axis=0), np.mean([d['percentage_iter'] for d in data_ls], axis=0), c=(0,0,0), linestyle='--', label='random')
plt.legend(loc="upper left")
```

```python
# PAINs recovery rate
for d in data_ls:
    plt.plot(d['percentage_iter'], d['pains_recovery_iter'], c=(0.8,0.8,0.8))
plt.plot(np.mean([d['percentage_iter'] for d in data_ls], axis=0), np.mean([d['pains_recovery_iter'] for d in data_ls], axis=0), c=(0,0,0), label='mean')
plt.plot(np.mean([d['percentage_iter'] for d in data_ls], axis=0), np.mean([d['percentage_iter'] for d in data_ls], axis=0), c=(0,0,0), linestyle='--', label='random')
plt.legend(loc="upper left")
```
